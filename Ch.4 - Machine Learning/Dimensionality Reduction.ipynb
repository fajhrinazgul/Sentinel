#### 1 - Exploring high dimensional data

Focus : Understanding t-Stochastic Neighbor Embedding Visualization for Feature Selection<br>
Dataset : Mushroom Classification Dataset

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

data = pd.read_csv("mushrooms.csv")
data.head()
```

```python
X = data.drop('class', axis=1)
y = data['class']

y = y.map({'p': 'Posionous', 'e': 'Edible'})
cat_cols= X.select_dtypes(include='object').columns.tolist()
for col in cat_cols:
    print (f"col name : {col}, N Unique : {X[col].nunique()}")
```

```python
for col in cat_cols:
    X[col]=X[col].astype("category")
    X[col]=X[col].cat.codes
X.head()
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.manifold import TSNE
X_std = StandardScaler().fit_transform(X)

tsne = TSNE(n_components=2)
X_tsne = tsne.fit_transform(X_std)
X_tsne_data = np.vstack((X_tsne.T, y)).T
df_tsne = pd.DataFrame(X_tsne_data, columns=['Dim1', 'Dim2', 'class'])
df_tsne.head()
```

```python
plt.figure(figsize=(8, 8))
sns.scatterplot(data=df_tsne, hue='class', x='Dim1', y='Dim2')
plt.show()
```
